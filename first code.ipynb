```python
from bs4 import BeautifulSoup
import requests
import csv
from itertools import zip_longest
```

```python
page_num = 1
site_list = {
                "shoptitle" :[],
                "prices"  : [],
                "areaa"  : [],
                "links" : []
                
            }
while True:

    if(page_num == 36):
        print("pages ended, terminate")
        break
        
    url= requests.get(f'https://sa.aqar.fm/%D9%85%D8%AD%D9%84%D8%A7%D8%AA-%D9%84%D9%84%D8%A5%D9%8A%D8%AC%D8%A7%D8%B1/%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6/%D8%B4%D8%B1%D9%82-%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6/{page_num}','utf8')
    shops= url.content
    soup= BeautifulSoup (shops,"lxml")
    
    shopstitles = soup.find_all('h4',{'class':'listTitle'})
    price= soup.find_all('span',{'class':'price'})
    area= soup.find_all('span',{'class':'size'})
    
    for i in range(len(shopstitles)):
        site_list["shoptitle"].append(shopstitles[i].text)
        site_list["links"].append(shopstitles[i].find('a',{'class':'listTitle'}).attrs['href'])
        site_list["prices"].append(price[i].text)
        site_list["areaa"].append(area[i].text)  
    
    page_num+= 1
    print("page switched to page "+ str(page_num))
```

```python
import pandas as pd
df = pd.DataFrame(site_list)
df
```

```python
df.to_csv("site_data1.csv", encoding="utf-8")
```
