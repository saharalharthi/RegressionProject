```python
from bs4 import BeautifulSoup
import requests
import csv
from itertools import zip_longest
import pandas as pd
import time
import numpy as np
```

```python
page_num = 1
East_Riyadh = {
                "shoptitle" :[],
                "prices"  : [],
                "areaa"  : [],
                "links"   :[]
                
                
            }
while True:

    if(page_num == 24):
        print("pages ended, terminate")
        break
        
    url= requests.get(f'https://sa.aqar.fm/%D9%85%D8%AD%D9%84%D8%A7%D8%AA-%D9%84%D9%84%D8%A5%D9%8A%D8%AC%D8%A7%D8%B1/%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6/%D8%B4%D8%B1%D9%82-%D8%A7%D9%84%D8%B1%D9%8A%D8%A7%D8%B6/{page_num}','utf8')
    shops= url.content
    soup= BeautifulSoup (shops,"lxml")
    
    shopstitles = soup.find_all('h4',{'class':'listTitle'})
    price= soup.find_all('span',{'class':'price'})
    area= soup.find_all('span',{'class':'size'})
    
    for i in range(len(shopstitles)):
        East_Riyadh["shoptitle"].append(shopstitles[i].text)
        East_Riyadh["prices"].append(price[i].text)
        East_Riyadh["areaa"].append(area[i].text)
        East_Riyadh["links"].append(shopstitles[i].find('a',{'class':'listTitle'}).attrs['href'])
   


        
        
    
    page_num+= 1
    print("page switched to page "+ str(page_num))
```

```python
import pandas as pd
dfeast = pd.DataFrame(East_Riyadh)
dfeast
```

```python
dfeast.to_csv("East_Riyadh.csv", encoding="utf-8")
```

```python
dfeast = pd.read_csv("East_Riyadh.csv")
dfeast
```

```python
soup_list3 = []
for index, l in enumerate(dfeast["links"]):
    newlink3 = requests.get("https://sa.aqar.fm" + l)
    #print(newlink.status_code)
    src3 = newlink3.text
    soup3 = BeautifulSoup(src3, "lxml")
    soup_list3.append(soup3)
    time.sleep(2)
```

```python
page_dict3 = {
    "front": [],
    "street_width":[],
    "age":[]
}
for ss in soup_list3:
    #print(ss.find('table',class_="dc-detailsTable").find_all("td")[0].text)
    page_dict3["front"].append(ss.find('table',class_="dc-detailsTable").find_all("td")[2].text)
    try:
        page_dict3["age"].append(ss.find('table',class_="dc-detailsTable").find_all("td")[6].text)
        page_dict3["street_width"].append(ss.find('table',class_="dc-detailsTable").find_all("td")[4].text)
    
    except:
        page_dict3["age"].append(np.nan)
        page_dict3["street_width"].append(np.nan)
sub_d3 = pd.DataFrame(page_dict3)
   
```

```python
df_merg_east= dfeast.merge(sub_d3, left_index=True, right_index=True)
df_merg_east
```

```python
df_merg_east.to_csv("final_east.csv", encoding="utf-8")
```
